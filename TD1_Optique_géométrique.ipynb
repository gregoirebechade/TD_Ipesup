```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
D = lambda i , A, n_prism : i + np.arcsin((n_prism * np.sin(A-np.arcsin(np.sin(i)/n_prism))))-A # calcul de l'angle de déviation en fonction de i, A et n_prism
x=np.arange(0, np.pi / 2 , 0.01)
A= 2*np.pi*30/360 # angle du prisme de 30° (attention à convertir en radians)
n_prism = 1.7 # indice du prisme 
y = D(x, A, n_prism)
plt.plot(x,y)
plt.title(r'$D(i)$')
plt.xlabel('D (radians)')
plt.ylabel('i (radians)')
plt.show()
```
